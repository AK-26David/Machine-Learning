# Matplotlib #

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

### 1.) Basic Graph ###

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Step 1 - Resize your graph 

plt.figure(figsize=(8,5),dpi=100) # figsize indicates the height and width of the graph 

# Step 2 - Keyword argument notation 

plt.plot(x, y, 'b^- -', label='2x') # Create a plot with blue triangle-up markers and dashed line, labeled '2x'#
```
